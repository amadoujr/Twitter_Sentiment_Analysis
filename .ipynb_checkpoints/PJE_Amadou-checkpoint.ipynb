```python
from tkinter import *
from tkinter import filedialog
from tkinter import ttk
import pandas as pd
import re
```

```python
class Application(Tk):
    def __init__(self):
        Tk.__init__(self)
        self.title("Tweets feelings")  
        self.minsize(800,760)  
        
    def uploadFile(self):
        fileName = filedialog.askopenfilename()
        self.cleanFile(fileName)
        
    def cleanFile(self,csvFile):
        file = pd.read_csv(csvFile)
        df = pd.DataFrame(file)
        oldColumns = df.columns # her, we have the information about the first row, pandas put it as a column
        df.columns = ['target','Ids','date','flag','user','text'] # renaming old columns 
        df.loc[len(df.index)] = oldColumns  # put the old column (first row) at the last row
        print(df.head())
        print(df.shape)
        print("----")
        self.processCleaning(df,"text")
        res = self.setPolarity(df)
        print(res.head())
    
    def removeSpecialCaractere(self,text):
        indice = 1
        while indice != 0:
            indice = 0
            copy = text
            #Replace @username with @
            text = re.sub(r'@[A-Za-z0-9]+', '@', text)
            #Delete #
            text = re.sub(r'#', '', text)
            #Replace $|£|€|% values with variable ($14.99 => $XX) (57€ => €XX)
            text = re.sub(r'[$£€%]\d+(\.\d+)?', lambda match: match.group(0)[0] + 'XX', text)
            text = re.sub(r'(\d+(\.\d+)?)[$£€%]', lambda match: match.group(0)[-1] + 'XX', text)
            #Delete RT
            text = re.sub(r'RT[\s]+', '', text)
            #Delete link
            text = re.sub('https?:\/\/\S+', '', text)
            #Adding space between every caracter preceded or followed by a ponctuation 
            text = re.sub(r'([.,:;?!*"()\[\]{}])([A-Za-z0-9])', r'\1 \2', text)
            text = re.sub(r'([A-Za-z0-9])([.,:;?!*"()\[\]{}])', r'\1 \2', text)
            #Delete emojis
            emoji_pattern = re.compile(r'[\U0001F600-\U0001F64F\U0001F300-\U0001F5FF\U0001F700-\U0001F77F\U0001F780-\U0001F7FF\U0001F800-\U0001F8FF\U0001F900-\U0001F9FF\U0001FA00-\U0001FA6F\U0001FA70-\U0001FAFF]+', flags=re.UNICODE)
            text = emoji_pattern.sub(r"", text)
            if text != copy:
                indice = 1
        return text
    
    def processCleaning(self,dataFrame,column):
        self.dropDuplicateElement(dataFrame,column)
        dataFrame[column] = dataFrame[column].apply(self.removeSpecialCaractere)        
        
    def dropDuplicateElement(self,dataFrame,column):
        dataFrame.drop_duplicates(subset=[column],inplace = True)
        
    def setPolarity(self,dataFrame):
        filePositive = open('positive.txt').read().splitlines()
        fileNegative = open('negative.txt').read().splitlines()
        countPositiveWord = 0
        countNegativeWord = 0
        dataFrame['polarity'] = -1
        
        for i, tweet in enumerate(dataFrame["text"]):
            res = tweet.split(' ')
            print(res)
            for word in res:
                if word in filePositive:
                    countPositiveWord+=1
                elif word in fileNegative:
                    countNegativeWord+=1
            if countPositiveWord > countNegativeWord:
                print('+++' ,countPositiveWord)
                dataFrame.at[i,'polarity'] = 4
            elif countPositiveWord < countNegativeWord:
                print('+++' ,countPositiveWord)
                dataFrame.at[i,'polarity'] = 2
            else:
                print('+++' ,countPositiveWord)
                dataFrame.at[i,'polarity'] = 0
        return dataFrame
            
        
    def createWidgets(self): 
        cadre_boutons = Frame(self) 
        cadre_boutons.pack(side = "top")
        choices = ["Option 1", "Option 2", "Option 3"]
        self.dropdown = ttk.Combobox(self, values=choices)
        self.dropdown.set("Select an option")
        self.uploadButton = Button(cadre_boutons, text = "Upload the learning base", command = self.uploadFile)
        self.uploadButton2 = Button(cadre_boutons, text = "Upload the tweets to classify", command = self.uploadFile)
        self.quitButton = Button(self, text = "Leave", command = self.destroy)
        self.dropdown.pack(side="right",anchor="ne")
        self.uploadButton.pack(side="left", padx=10)
        self.uploadButton2.pack(side="right", padx=10)
        self.quitButton.pack(side="bottom",anchor="se")
        
app = Application()
app.createWidgets()
app.mainloop()
```

```python
 def removeSpecialCaractere(text):
        #1, Replace @username with @
    text = re.sub(r'(@[a-zA-Z0-9]+)', '@', text)
        #2, Add white space character before and after ! ? : . " ; ,
    text = re.sub(r'([!?":;,])', r' \1 ', text)
        #3, Replace dollar values with variable ($14.99 => $XX)
    text = re.sub(r'(\$|€|£)([\d.]+)', r'\1XX', text)
    
    #4, Replace percentage values with variable (25n% => XXn%)
    text = re.sub(r'(\d+)%', r'XX%', text)    
    return text

        
```
